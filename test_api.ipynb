```python
import pandas as pd
import numpy as np
import requests
```

### Coverage Search

```python
zip_list = [80863]
```

```python
zip_provider_list = []

for curr_zip in zip_list:
    response = requests.get('https://marketplace.api.healthcare.gov/api/v1/coverage/search?'\
                            'apikey=sPB0vYExMK1Qj730zOyl2Mdh53tmGhNd'\
                            '&q=health system'\
                            '&zipcode={}'.format(curr_zip))

    zip_providers = response.json()['providers']

    for x in range(len(zip_providers)):
        # provider info
        provider_json = zip_providers[x]['provider']
        provider_npi = provider_json['npi']
        provider_name = provider_json['name']
        provider_type = provider_json['facility_types']
        provider_taxonomy = provider_json['taxonomy']

        # address info
        address_json = zip_providers[x]['address']
        provider_street = address_json['street1']
        provider_zip = address_json['zipcode']
        provider_phone = address_json['phone']

        # distance info
        provider_distance_from_zip = round(zip_providers[x]['distance'], 2)

        dict_provider = {}
        dict_provider.update({'Zip':curr_zip,
                            'Provider_NPI':provider_npi,
                            'Provider_Name':provider_name, 
                            'Provider_Type':provider_type, 
                            'Provider_Taxonomy':provider_taxonomy,
                            'Provider_Street':provider_street, 
                            'Provider_Zip':provider_zip, 
                            'Provider_Phone':provider_phone, 
                            'Provider_Distance_fromZip':provider_distance_from_zip})

        zip_provider_list.append(dict_provider)
 
```

```python
provider_df = pd.DataFrame(zip_provider_list) 
```
